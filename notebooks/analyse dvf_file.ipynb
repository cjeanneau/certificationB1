```python
# On charge le fichier dans un dataframe pandas
df_ori = pd.read_csv(file_path, sep='|', dtype=str)
```

```python
# On élimine les colonnes totalement vides
df = df_ori.dropna(axis=1, how='all')
df.shape
```

```python
# Les échanges ne nous intéressent pas ici
print(f"Nb lignes avant nettoyage : {df.shape}")
df2 = df[df['Nature mutation'] != 'Echange']
print(f"Nb lignes après nettoyage : {df2.shape}")
```

```python
df_with_No_Volume = df2[df2['No Volume'].notna()]
df_with_No_Volume.describe()
```

```python
# On remarque ici que les lignes contenant un No Volume non vide n'ont aucune surface précisée (construite ou terrain).
# Elles ne nous servent donc à rien 
print(f"Nb lignes avant nettoyage : {df2.shape}")
df3 = df2[df2['No Volume'].isna()]
print(f"Nb lignes après nettoyage : {df3.shape}")
# Puis on efface cette colonne qui ne contient plus que des Nan
df4 = df3.drop('No Volume', axis=1)
print(f"Nb lignes après nettoyage : {df4.shape}")
```

```python
df4.describe()
```

```python
# on va sélectionner les colonnes à garder puisque des colonnes comme lot et surface carrez ne nous interesse pas, 
# Nombre de lots semble êter renseigné aléatoirement.
# On pourra reconstruire l'adresse = 'No voie' + 'B/T/Q' + 'Type de voie' + 'Voie' + 'Code postal' + 'Commune'
# On pourra reconstruire le code insee commune = 'Code departement' +'Code commune'
# On pourra reconstruire la référence cadastrale = 'Prefixe de section' + 'Section' +'No plan'
# On garde les Nature de culture pour le moment car elle différencie les types de de terrain lors d'une même vente.

col_to_keep = [
    'Date mutation',
    'Nature mutation',
    'Valeur fonciere',
    'No voie',
    'B/T/Q',
    'Type de voie',
    'Code voie',
    'Voie',
    'Code postal',
    'Commune',
    'Code departement',
    'Code commune',
    'Prefixe de section',
    'Section',
    'No plan',
    'Code type local', 
    'Type local', 
    'Surface reelle bati', 
    'Nombre pieces principales',
    'Nature culture',
    'Nature culture speciale',
    'Surface terrain'
    ]
df5 = df4[col_to_keep]
df5.describe()
```

```python
# On va aussi supprimer less lignes dont la valeur foncière est Nan
df6 = df5[df5['Valeur fonciere'].notna()]
df6.describe()
```

```python
# on élimine les +30000 lignes avec des valeurs foncières = 1,00 
df7 = df6[df6['Valeur fonciere'] != '1,00']
df7.describe()
```

```python
df7.loc[:, 'Valeur fonciere'] = df7['Valeur fonciere'].str.replace(',', '.', regex=False).astype(float)
```

```python
df7.head()
```

```python
print(f"\nNombre de lignes en double avant suppression : {df7.duplicated().sum()}")

# Supprimer les doublons (conserver la première occurrence)
df8 = df7.drop_duplicates()

print(f"Nombre de lignes en double après suppression : {df8.duplicated().sum()}")
print(f"Nombre de lignes après suppression des doublons : {df8.shape[0]}")
```

```python
df8.head()
```

```python
#Petit ménage dû à quelques ralentissements
## J'aurais dû optimiser mes typs de données dès le début !
%whos
```

```python
del df, df2, df3, df4, df5, df6, df_with_No_Volume
import gc
gc.collect()
```

```python
# On va remplacer les valeurs Nan de 'Surface reelle bati' et 'Surface terrain' par '0'

print("Nombre de NaN dans 'Surface reelle bati' avant :", df8['Surface reelle bati'].isnull().sum())
print("Nombre de NaN dans 'Surface terrain' avant    :", df8['Surface terrain'].isnull().sum())

df8.loc[:, 'Surface reelle bati'] = df8['Surface reelle bati'].fillna('0')
df8.loc[:, 'Surface terrain'] = df8['Surface terrain'].fillna('0')

print("\nNombre de NaN dans 'Surface reelle bati' après :", df8['Surface reelle bati'].isnull().sum())
print("Nombre de NaN dans 'Surface terrain' après    :", df8['Surface terrain'].isnull().sum())
```

```python
# et maintenant on supprime les lignes dont les 2 variables de surface (terrain et bati) sont nulles
filtre = ((df8['Surface reelle bati'] == '0') & (df8['Surface terrain'] == '0'))
print("Nombre de lignes à effacer : ", filtre.sum())
print("Nombre de ligne du df avant nettoyage : ", len(df8))
df8 = df8[~filtre]
print("Nombre de ligne du df après nettoyage : ", len(df8))

```

```python
pd.set_option('display.max_rows', 100)
```

```python
filtre = ((df8['Surface reelle bati'] == '0') & (df8['Surface terrain'] != '0'))
print(filtre.sum())
```

```python
# Nous alllons regrouper le stransaction par date et valeurs et créer pour chacun de ces couples un id_transaction
df8['Date mutation'] = pd.to_datetime(df8['Date mutation'], format='%d/%m/%Y', errors='coerce') # on s'assure que toutes les dates soient au même format pour le tri, c'est mieux !
df8_sorted = df8.sort_values(by=['Date mutation', 'Valeur fonciere', 'Code departement', 'Code commune']).reset_index(drop=True)
df8_sorted['id_transaction'] = df8_sorted.groupby(['Date mutation', 'Valeur fonciere', 'Code departement', 'Code commune', 'Code voie']).ngroup()
df8_sorted.loc[:,'id_transaction'] = df8_sorted['id_transaction'].fillna(0)
df8_sorted['id_transaction'] = df8_sorted['id_transaction'].astype(int)
```

```python
df8_sorted[df8_sorted['Surface reelle bati'] != '0'].head(10)
```

```python
# Filtre des id_transation des biens ayant une surface batie
id_surface_bati = df8_sorted[df8_sorted['Surface reelle bati'] != '0']['id_transaction']
print(id_surface_bati)
```

```python
id_surface_bati_list = id_surface_bati.tolist()
transaction_avec_bati = df8_sorted[df8_sorted['id_transaction'].isin(id_surface_bati_list)].sort_values(by='id_transaction')
pd.set_option('display.max_rows', 100)
transaction_avec_bati[['id_transaction', 'Date mutation', 'Valeur fonciere', 'Code departement', 'Code commune', 'Code voie', 'Type local', 'Surface reelle bati', 'Surface terrain']].head(100)
```

```python
# On remarque ici que tous les types de locaux Nan n'ont pas de surface bati.

transaction_avec_bati[(transaction_avec_bati['Type local'].isna()) & (transaction_avec_bati['Surface reelle bati'] != '0')]
```

```python
# De même por les dépendances qui ne concernent que des surfaces de terrain.
transaction_avec_bati[(transaction_avec_bati['Type local'] == 'Dépendance') & (transaction_avec_bati['Surface reelle bati'] != '0')]
```

```python
# Le prix d'un terrain étant insignifiant dans le prix d'un bien immobilier, nous allons supprimer ces lignes.
filtre1 = transaction_avec_bati['Type local'].isna()
filtre2 = transaction_avec_bati['Type local'] == 'Dépendance'
print(filtre1.sum())
print(filtre2.sum())
filtre = filtre1 | filtre2
print(filtre.sum())
print(len(transaction_avec_bati))
transaction_avec_bati_uniquement = transaction_avec_bati[~filtre]
print(len(transaction_avec_bati_uniquement))
```

```python
transaction_avec_bati_uniquement[['id_transaction', 'Date mutation', 'Valeur fonciere', 'Code departement', 'Code commune', 'Code voie', 'Type local', 'Surface reelle bati', 'Surface terrain']].head(40)
```

```python
# Tester si on ne garde que les lignes uniques ?
# Car si on prend le risque d'additionner des surfaces, nous ne sommes pas certains que ce n'est pas une mise à jour des données 
# et donc cela fausserait notre dataset

id_transaction_count = transaction_avec_bati_uniquement['id_transaction'].value_counts()
unique_id_transactions_values = id_transaction_count[id_transaction_count == 1].index
unique_id_transactions_list = unique_id_transactions_values.tolist()

df =  transaction_avec_bati_uniquement[transaction_avec_bati_uniquement['id_transaction'].isin(unique_id_transactions_list)]
df_inv = transaction_avec_bati_uniquement[~transaction_avec_bati_uniquement['id_transaction'].isin(unique_id_transactions_list)]
```
